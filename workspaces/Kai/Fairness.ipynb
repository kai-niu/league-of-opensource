```python
from sklego.metrics import equal_opportunity_score
from sklearn.linear_model import LogisticRegression
import types
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import pickle
```

```python
df = pd.read_csv('../../assets/dataset/relocation_db_with_gender.csv', index_col=None)
# X,y
y = df.loc[:,'life_allocation']
X = df.drop('life_allocation', axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
model = pickle.load(open("life_allocation_gender.pkl", "rb"))
fairness = equal_opportunity_score(sensitive_column="GENDER")(model, X_test, y_test)
print('equal_opportunity_score:', fairness)
```
