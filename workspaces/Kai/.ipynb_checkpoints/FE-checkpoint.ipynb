```python
import pandas as pd
import numpy as np
```

```python
customer = pd.read_csv('../../assets/dataset/customer.csv', index_col=None)
event = pd.read_csv('../../assets/dataset/event.csv', index_col=None)
```

## Derive Labels
* target event: LFE_HOME_PURCHASE
* target event: LFE_RELOCATION

```python
event.head()
```

```python
event['home_purchase'] = event['EVENT_TYPE_ID'] == 'LFE_HOME_PURCHASE'
event['life_allocation'] = event['EVENT_TYPE_ID'] == 'LFE_RELOCATION'
```

```python
labels = event[['CUSTOMER_ID','EVENT_DATE','home_purchase','life_allocation']]
labels['EVENT_DATE'] = pd.to_datetime(event['EVENT_DATE'], format='%Y-%M-%d')
labels.head()
```

## FE

```python
# null rate
foo = customer.isna().sum()/customer.shape[0]
foo[foo < .7]
df1=customer.loc[:, foo.index]
df1.head()
```

```python
##
# keep numeric value only
keeps = []
dtypes = df1.dtypes
for i,t in enumerate(dtypes):
    if str(t) == "object":
        continue
    else:
        keeps.append(dtypes.index[i])
df2 = df1.loc[:, keeps]
df2['EFFECTIVE_DATE'] = pd.to_datetime(df1['EFFECTIVE_DATE'], format='%Y-%M-%d')
df2 = df2.sort_values(['CUSTOMER_ID', 'EFFECTIVE_DATE'])
df2.head()
```

## Merge Dataset

```python
final = pd.merge(df2, labels, how='right',
                 left_on=["CUSTOMER_ID"], 
                      right_on=["CUSTOMER_ID"])

final = final.loc[final['EVENT_DATE'] > final['EFFECTIVE_DATE'], :]
print(final.shape, df2.shape)
final.head()
```

```python
home_db = final.drop(['EFFECTIVE_DATE','EVENT_DATE','life_allocation'], axis=1)
relocation_db = final.drop(['EFFECTIVE_DATE','EVENT_DATE','home_purchase'], axis=1)
```

## Save DB

```python
home_db.to_csv('../../assets/dataset/home_db.csv')
relocation_db.to_csv('../../assets/dataset/relocation_db.csv')
```
